```python
# FILE_NAME = "1. vlna všetko new 13.11.2024.xlsx"
# FILE_NAME = "2. vlna všetko new 13.11.2024.xlsx"
# FILE_NAME = "3. vlna všetko new 13.11.2024.xlsx"
FILE_NAME = "4. vlna do 31.5.2024 všetko new.xlsx"
N = 4 # vlna
SOURCE_DIR = "../source_data/"
```

```python
import os.path

import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_excel(os.path.join(SOURCE_DIR, FILE_NAME))
# drop unnamed columns
df = df.loc[:, ~df.columns.str.contains("^Unnamed")]
# drop columns: Kluc, Poisťovňa, Záznam
df.drop(columns=["Kluc", "Meno", "Poisťovňa", "Záznam", "S-Bil-D"], inplace=True)
df.head()
```

```python
print("Length of the data:", len(df))
```

```python
# dátum príjmu a dátum prepustenia by mali byť dátumovými typmi
df["Dátum príjmu"] = pd.to_datetime(df["Dátum príjmu"], format="%d.%m.%Y")
df["Dátum prepustenia"] = pd.to_datetime(df["Dátum prepustenia"], format="%d.%m.%Y")
# Pohlavie by malo byť kategorickým typom
df["Kód prepustenia"] = df["Kód prepustenia"].astype("int")
df = df[df["Kód prepustenia"].isin([1, 2, 3, 4, 5, 6, 7])]
df["Kód prepustenia"] = df["Kód prepustenia"].astype("category")
df["Pohlavie"] = df["Pohlavie"].astype("category")
df["Kód príjmu"] = df["Kód príjmu"].astype("category")
df["HLN Dg."] = df["HLN Dg."].astype("category")
df.dtypes
# drop rows where "Kód prepustenia" is not Category 1, 2, 3, 4, 5, 6 or 7
```

```python
plt.rcParams.update({
    'axes.titlesize': 18,
    'axes.labelsize': 14,
    'xtick.labelsize': 14,
    'ytick.labelsize': 14,
    'legend.fontsize': 12
})
```

```python
# visualize age distribution using column "Poradie"
plt.figure(figsize=(10, 6))
sns.histplot(df["Vek"], kde=True)
plt.title(f"Distribúcia veku {N}. vlny")
plt.xlabel("Vek")
plt.ylabel("Frekvencia")
plt.grid()
plt.savefig(f"../images/vek_{N}.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Vytvorenie počtového grafu podľa pohlavia
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x="Pohlavie")

# Nastavenie názvov a veľkostí písma
plt.title(f"Počet prípadov podľa pohlavia {N}. vlny", fontsize=16)
plt.xlabel("Pohlavie", fontsize=14)
plt.ylabel("Počet", fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Uloženie grafu
plt.savefig(f"../images/pohlavie_{N}.png", dpi=300, bbox_inches="tight")

# Zobrazenie grafu
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x="Kód prepustenia", order=sorted(df["Kód prepustenia"].dropna().unique()))

plt.title(f"Počet prípadov podľa kódu prepustenia – {N}. vlna", fontsize=16)
plt.xlabel("Kód prepustenia", fontsize=14)
plt.ylabel("Počet", fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Uloženie obrázku
plt.savefig(f"../images/kod_prepustenia_{N}.png", dpi=300, bbox_inches="tight")

plt.show()
```

```python
df.groupby(df["Dátum príjmu"].dt.to_period("M")).size().plot(
    kind="line", figsize=(10, 5), label="Prijatý", color="red"
)
df.groupby(df["Dátum prepustenia"].dt.to_period("M")).size().plot(
    kind="line", figsize=(10, 5), label="Prepustený", color="green"
)
plt.title(f"Počet prepustených a prijatých pacientov za dátum {N}. vlny", size=15)
plt.ylabel("Počet pacientov")
plt.xlabel("Dátum")
plt.legend()
plt.grid(True)
plt.savefig(f"../images/prijati_prepusteni_{N}.png", dpi=300, bbox_inches="tight")
```
